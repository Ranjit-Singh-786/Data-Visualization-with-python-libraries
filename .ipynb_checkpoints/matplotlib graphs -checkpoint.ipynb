# SUBPLOTS
## with the help of subplots function you can draw the multiple chart in a single in a single figure

# subplot(1,2)
here first parameter difines the row and second is colomn and third parameter is  defines the number of chart

```python
#first plot 
x=np.array([51,81,90,75,59])
y=np.array([21,13,25,41,31])
plt.subplot(1,2,1)          # 1 row and 2 column and 3rd denotes the plot no
plt.plot(x,y)
plt.title('first chart')
plt.grid(axis='x',color="k")
plt.xlabel('( 1-2-1 )')
# second plot
plt.subplot(1,2,2)
x2=np.array([2,5,4,8,7])
y2=np.array([8,4,5,9,6])
plt.plot(x2,y2)
plt.title('second chart')
plt.xlabel('( 1-2-2 )')
print('here i am draw a 1 row and 2 columns ,next parameter defines whose number of chart')
```

# subplot(2,1)

```python
#first plot 
x=np.array([51,81,90,75,59])
y=np.array([21,13,25,41,31])
plt.subplot(2,1,1)      # row , column , chart no.
plt.plot(x,y)
plt.title('first chart')
plt.grid(axis='x',color="k")
plt.ylabel('( 1-2-1 )')
# second plot

plt.subplot(2,1,2)
x2=np.array([2,5,4,8,7])
y2=np.array([8,4,5,9,6])
plt.plot(x2,y2)
plt.title('second chart')
plt.xlabel('( 1-2-2 )')
```

# wedgeprops
<p> is used for the slices property</p>

```python
# ,,,,,,,,,,,,,wedgeprops----->edgecolor=black
plt.figure(figsize=(12,7))
slices=[412,215,367,521,254]
labell=['python','java','java script','c','c++']

plt.pie(slices,labels=labell,shadow=True,wedgeprops={'edgecolor':'black','linewidth':2,'linestyle':'-'},autopct="%1.2f")
plt.show()
```

#   startangle
<p> is used for start the representation of slices</p>
<br>
<p>in this parameter we have only 4 startangle and there are</p><br>
<ul>
    <li>0   degree angle</li>
    <li>90  degree angle</li>
    <li>180 degree angel</li>
    <li>270 degree agel</li>
    </ul>

```python
# ,,,,,,,,,,startagnle=90,,,,,,,,0,180,90,270 there are four startangle
plt.figure(figsize=(12,7))

slices=[412,215,367,521,254]
labell=['python','java','java script','c','c++']

plt.pie(slices,labels=labell,wedgeprops={'ec':'k'},startangle=90)

# you can not define the ec outside the wedgeprops
plt.show()
```

# autopct
<p> is used for show the percentage value for each slices</p>
paramater is <b>"%1.1f%%"</b>
<p> here you can use 1integer value and 2f it means how many floating number you want to see</p>

```python
# ,,,,,,,,,,,,autopct,,,,,,,,,,,,
plt.figure(figsize=(10,7))
slices=[412,215,367,521,254]
labell=['python','java','java script','c','c++']

plt.pie(slices,labels=labell,wedgeprops={'edgecolor':'black'},shadow=True,startangle=0,autopct="%1.1f%%")
plt.show()
```

# code of this chart access from the greatlearning
<P> i extracted this coaded for the know more about of pie chart...</P>

```python
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD', 
        'TESLA', 'JAGUAR', 'MERCEDES']
  
data = [23, 17, 35, 29, 12, 41]
  
  
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)
  
# Creating color parameters
colors = ( "orange", "cyan", "brown",
          "grey", "indigo", "beige")
  
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "green" }
  
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
  
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data, 
                                  autopct = lambda pct: func(pct, data),      #awesome trick lambda with in function
                                  explode = explode, 
                                  labels = cars,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))
  
# Adding legend
ax.legend(wedges, cars,
          title ="Cars",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
  
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")
plt.show()    # if you are using show()then you can watch only chart no other any text
  
```

# savefigure function
<p> this function is used for when will you want to save the figure,<br>
    you can use the <b>savefigure function</b> and in this function you can <br>
    putt the multiple parameter. like facecolor,path or filename,dpi,format,,,etc..</p>

```python
plt.figure(figsize=(9,6))
col=['r'if i%2==0 else 'w' for i in range(1,21)]
col2=['w'if i%2==0 else 'b' for i in range(1,21)]
col3=['r','w','b','brown']
col4=['k'if i%2==0 else 'w' for i in range(1,11)]
inside=np.ones(10)
lab2=np.ones(20)
lab=np.ones(20)
plt.pie([1],colors='k',radius=2.05)
plt.pie(lab,colors=col,radius=2.0)
plt.pie([1],colors='k',radius=1.1)
plt.pie([1],colors='g',radius=1.5)
plt.pie(lab2,colors=col2,radius=1.07)
plt.pie([1],colors='m',radius=0.7)
plt.pie(inside,colors=col4,radius=0.3)
plt.pie([1,1,1,1],colors=col3,radius=0.1)
plt.savefig("pie_chart",facecolor='c')      # you  can use multiple parameter here
plt.show()
```

# stackplot 
<p>stackplot is also called area plot this plot is very interesting <br>
    and more helpfull for know the about data distubution and <br>
    and in this chart you can also draw the multiple area in a single plot</p>

```python
plt.figure(figsize=(9,6))
style.use('seaborn-dark')
plt.subplot(3,1,1)
plt.stackplot(minute,player1,color='r',labels='player1')
plt.legend(loc='upper left')
plt.grid(axis='y',color='k')
# plt.title('player_1')
plt.subplot(3,1,2)
plt.stackplot(minute,player2,labels='player2')
plt.legend(loc='upper left')

# plt.title('player_2')
plt.subplot(3,1,3)
plt.stackplot(minute,player3,color='g',labels='player3')
plt.legend(loc='upper left')
# plt.title('player_3')
plt.show()
```

# and now we will draw the stackplot of tha above mentione three chart in a sinlge stackplot
<p> then you can look a interesting fact of data</p>
<B>NESTED STACKPLOT </B>

```python
style.use('classic')
plt.figure(figsize=(8,7))
minute=[i for i in range(1,10)]
player1=[1, 2, 2, 2, 3, 4, 5, 5, 5]
player2=[2, 4, 4, 5, 5, 5, 7, 8, 8]
player3=[1,4,7,5,4,1,2,4,5]
labl=['player1','player2','player3']
plt.stackplot(minute,player1,player2,player3,labels=labl)
plt.legend(loc='upper left')
plt.title('stackplot')
plt.xlabel('minutes',fontsize=15)
plt.ylabel('players values',fontsize=15,color='b',family='serief')
plt.grid(color='k',linewidth=1,linestyle='solid')
plt.show()
print('here y axis is representing the total score of players by index or you can say ')
print('there are all three player are ziped and total value representing a stackplot')
```
