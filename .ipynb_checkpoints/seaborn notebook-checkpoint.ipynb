```python
sns.relplot(x='time',y='firing_rate',data=dot,kind='line',size='coherence',sizes=(1,5),hue='coherence',style='choice',palette=['r','g','y','b','c','k'])
```

# this chart tricks is very usefull in this trick i am using allmost parameter of relplot ()

```python
sns.relplot(x='time',y='firing_rate',data=dot,kind='line',lw=2,hue='coherence',style='choice',col='coherence',col_wrap=2,height=3,palette=['r','g','y','b','c','k'])
plt.title('data visulization trick')
# here size parameter is not returning best result ,hence here i am using lw-->linewidth
```
