# relplot
<p>remebered that you can also use pyplot property here</p>

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker')
```

# col parameter

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker',row='time')
```

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker',col='time')
plt.grid(axis='y')
# hue parameter change the color according to category of a columns
```

# style parameter in relplot()

```python
sns.relplot(x='total_bill',y='tip',data=data,hue='smoker',style='time')
# these plot is very helpfull for the particular point
print('and now i will seprate lunch and dinner records in the scatterplot ')
print('by using col parameter')
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='time',col='time',data=data)
print('u can see here now lunch and dinner point has seprated in two columns')
```

```python
sns.relplot(x='total_bill',y='tip',hue='size',data=data,style='time')

print('here available category in the size columns is bigger size')
print('so color combination divided in to maximum color and separete color represent a ')
print('separete catogory')
```

# color parameter for chang the single color in proper points

# for change the size with size parameter

```python
sns.relplot(x='tip',y='total_bill',data=data,size='size')
plt.show()
print('here you can see now size are changing according to ')
print(' my size value ')
```

# for change defualt size of size parameter with the help of sizes parameter 

```python
sns.relplot(x='total_bill',y='tip',data=data,size='size')
print('this is defualt size but  we will customized the defualt size')
```

```python
sns.relplot(x='total_bill',y='tip',color=['r'],data=data,size='size',sizes=(100,500))
print('this size is changin in between the 100 to 500')
```

```python
df=pd.DataFrame(np.random.randn(500,2).cumsum(axis=0),columns=['time','values'])
df.head()
```

```python
sns.relplot(x='time',y='values',kind='line',data=df,height=4)
```

```python
sns.relplot(x='time',y='values',kind='line',data=df,height=4,sort=False)
print('when i did sort false then you can see the result')
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
fmri.info()
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri)
print('you can see here confident interval')
```

# ci=False for removing the confident interval

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,ci=False)
```

# pass standard division in ci parameter

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,ci='sd')
print('it is returning a standard division chart with confident interval')
```

# here estimator used for create a chart with actual value

```python
sns.relplot(x='timepoint',y='signal',estimator=None,kind='line',data=fmri)
```

# hue chart two lines bcz there are category

```python
import matplotlib.style as slt
slt.use('ggplot')
sns.relplot(x='timepoint',y='signal',hue='event',kind='line',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',hue='region',style='event',kind='line',data=fmri)
```

# without confident interval but marker and  dashes parameter

```python
sns.relplot(x='timepoint',y='signal',hue='region',style='event',kind='line',data=fmri,markers=True,ci=None,dashes=False)
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',hue='event',style='event',data=fmri,markers=True,ci=None,palette=['g','b'])
```
